```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))

# Create scatter plot
scatter = plt.scatter(
    x=range(len(df)),
    y=df['waiting'],
    c=df['eruptions'],
    cmap='viridis',
    s=100,  # size of the points
    alpha=0.7  # transparency
)

# Add color bar to show the gradient of eruption intensity
plt.colorbar(scatter, label='Eruption Intensity')

# Set axis labels and title
plt.xlabel('Eruption Count')
plt.ylabel('Waiting Time (minutes)')
plt.title('Eruption Count vs Waiting Time with Eruption Intensity Gradient')

# Show the plot
plt.show()
```

```python
# Group by 'eruptions' and 'waiting' and count occurrences
grouped_df = df.groupby(['eruptions', 'waiting']).size().reset_index(name='counts')
plt.figure(figsize=(10, 6))

# Create bubble scatter plot
scatter = plt.scatter(
    y=grouped_df['eruptions'],
    x=grouped_df['waiting'],
    s=grouped_df['counts'] * 100,  # scale the bubble size
    c=grouped_df['counts'],        # color based on counts
    cmap='viridis',
    alpha=0.6
)

# Add color bar to show the count gradient
plt.colorbar(scatter, label='Count of Occurrences')

# Set axis labels and title
plt.ylabel('Eruption Intensity')
plt.xlabel('Waiting Time (minutes)')
plt.title('Bubble Scatter Plot of Eruptions vs Waiting Time')

# Show the plot
plt.show()
```
